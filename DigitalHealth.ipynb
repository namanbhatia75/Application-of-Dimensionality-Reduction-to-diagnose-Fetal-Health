# Imports


```python
import pandas as pd
```

```python
df = pd.read_csv("/content/sample_data/fetal_health_project2.csv")
df.head()
```

```python
Y = df['fetal_health']

Y
```

```python
X = df.drop(columns=['fetal_health'])
X
```

# Preprocessing

```python
X.isnull().any()
```

```python
Y.isnull().any()
```

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, trainy, testy = train_test_split(X, Y, test_size=0.30, random_state=42)
trainy
```

# Logistic Regression

Given an input variable, logistic regression is the process of modeling the probability of a discrete outcome. 



```python
from sklearn.linear_model import LogisticRegression
logisticRegr = LogisticRegression(solver='lbfgs', max_iter=10000)
logisticRegr.fit(train_X, trainy)
```

```python
predictions = logisticRegr.predict(test_X)
scoreLR = logisticRegr.score(test_X, testy)
print(scoreLR)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
```

```python
cm = metrics.confusion_matrix(testy, predictions)
print(cm)
```

```python
plt.figure(figsize=(3,3))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Pastel1');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(scoreLR)
plt.title(all_sample_title, size = 15);
```

```python
from sklearn.metrics import f1_score
f1_score(testy, predictions, average='weighted')
```

# Neural Network


```python
import tensorflow as tf
import numpy as np
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.InputLayer(input_shape=(train_X.shape[1], )), #Input Layer
    tf.keras.layers.Dense(1024, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(4, activation='softmax') #Output layer
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])
```

```python
history = model.fit(train_X, trainy, validation_split=0.30, epochs=5, verbose=0)
```

```python
plt.plot(history.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
model.evaluate(test_X, testy)
```

```python
neuralPredict = model.predict(test_X)
```

```python
# f1_score(testy, neuralPredict, average=None)
from sklearn.metrics import classification_report
y_pred_bool = np.argmax(neuralPredict, axis=1)
print(classification_report(testy, y_pred_bool))
```

# Feature Selection

```python
#Threshold Variance
from sklearn.feature_selection import VarianceThreshold
vt = VarianceThreshold(threshold=1)
train_X_reduced = vt.fit_transform(train_X)
test_X_reduced = vt.transform(test_X) 
print(train_X.shape)
print(train_X_reduced.shape)
```

```python
# Logistic Regression
logisticRegr.fit(train_X_reduced, trainy)
LRpred = logisticRegr.predict(test_X_reduced)
FSscoreLR = logisticRegr.score(test_X_reduced, testy)
print("Logistic Regression :: accuracy = ",FSscoreLR)
print("LR :: f1 = ", f1_score(testy, LRpred, average='weighted'))

# Naive Bayes
NBmodel.fit(train_X_reduced,trainy)
NBpred = NBmodel.predict(test_X_reduced)
FSscoreNB = NBmodel.score(test_X_reduced, testy)
print("Naive Bayes :: accuracy = ",FSscoreNB)
print("NB :: f1 = ", f1_score(testy, NBpred, average='weighted'))

# KNN
neigh.fit(train_X_reduced, trainy)
KNNpred = neigh.predict(test_X_reduced)
FSscoreKNN = neigh.score(test_X_reduced, testy)
print("K Nearest Neighbours :: accuracy = ",FSscoreKNN)
print("KNN :: f1 = ", f1_score(testy, KNNpred, average='weighted'))

# Decsision Tree
DT.fit(train_X_reduced, trainy)
DTpred = DT.predict(test_X_reduced)
FSscoreDT = DT.score(test_X_reduced, testy)
print("Decision Tree :: accuracy = ",FSscoreDT)
print("DT :: f1 = ", f1_score(testy, DTpred, average='weighted'))

#Support Vector Machine
SV.fit(train_X_reduced, trainy)
SVpred = SV.predict(test_X_reduced)
FSscoreSV = SV.score(test_X_reduced, testy)
print("Support Vector Machine :: accuracy = ",FSscoreSV)
print("SVM :: f1 = ", f1_score(testy, SVpred, average='weighted'))
```

```python
#ANOVA
from sklearn.feature_selection import SelectKBest,f_classif
skb = SelectKBest(f_classif, k=20).fit(train_X, trainy)
train_X_select = skb.transform(train_X)
test_X_select = skb.transform(test_X)
train_X_select.shape
```

```python
# Logistic Regression
logisticRegr.fit(train_X_select, trainy)
LRpred = logisticRegr.predict(test_X_select)
ANscoreLR = logisticRegr.score(test_X_select, testy)
print("Logistic Regression :: accuracy = ",ANscoreLR)
print("LR :: f1 = ", f1_score(testy, LRpred, average='weighted'))

# Naive Bayes
NBmodel.fit(train_X_select,trainy)
NBpred = NBmodel.predict(test_X_select)
ANscoreNB = NBmodel.score(test_X_select, testy)
print("Naive Bayes :: accuracy = ",ANscoreNB)
print("NB :: f1 = ", f1_score(testy, NBpred, average='weighted'))

# KNN
neigh.fit(train_X_select, trainy)
KNNpred = neigh.predict(test_X_select)
ANscoreKNN = neigh.score(test_X_select, testy)
print("K Nearest Neighbours :: accuracy = ",ANscoreKNN)
print("KNN :: f1 = ", f1_score(testy, KNNpred, average='weighted'))

# Decsision Tree
DT.fit(train_X_select, trainy)
DTpred = DT.predict(test_X_select)
ANscoreDT = DT.score(test_X_select, testy)
print("Decision Tree :: accuracy = ",ANscoreDT)
print("DT :: f1 = ", f1_score(testy, DTpred, average='weighted'))

#Support Vector Machine
SV.fit(train_X_select, trainy)
SVpred = SV.predict(test_X_select)
ANscoreSV = SV.score(test_X_select, testy)
print("Support Vector Machine :: accuracy = ",ANscoreSV)
print("SVM :: f1 = ", f1_score(testy, SVpred, average='weighted'))


yLR.append(ANscoreLR)
yNB.append(ANscoreNB)
yKNN.append(ANscoreKNN)
yDT.append(ANscoreDT)
ySVM.append(ANscoreSV)
```

```python
# Logistic Regression
logisticRegr.fit(PCAtrainX, trainy)
LRpred = logisticRegr.predict(PCAtestX)
scoreLR = logisticRegr.score(PCAtestX, testy)
print("Logistic Regression :: accuracy = ",scoreLR)
print("LR :: f1 = ", f1_score(testy, LRpred, average='weighted'))

# Naive Bayes
NBmodel.fit(PCAtrainX,trainy)
NBpred = NBmodel.predict(PCAtestX)
scoreNB = NBmodel.score(PCAtestX, testy)
print("Naive Bayes :: accuracy = ",scoreNB)
print("NB :: f1 = ", f1_score(testy, NBpred, average='weighted'))

# KNN
neigh.fit(PCAtrainX, trainy)
KNNpred = neigh.predict(PCAtestX)
scoreKNN = neigh.score(PCAtestX, testy)
print("K Nearest Neighbours :: accuracy = ",scoreKNN)
print("KNN :: f1 = ", f1_score(testy, KNNpred, average='weighted'))

# Decsision Tree
DT.fit(PCAtrainX, trainy)
DTpred = DT.predict(PCAtestX)
scoreDT = DT.score(PCAtestX, testy)
print("Decision Tree :: accuracy = ",scoreDT)
print("DT :: f1 = ", f1_score(testy, DTpred, average='weighted'))

#Support Vector Machine
SV.fit(PCAtrainX, trainy)
SVpred = SV.predict(PCAtestX)
scoreSV = SV.score(PCAtestX, testy)
print("Support Vector Machine :: accuracy = ",scoreSV)
print("SVM :: f1 = ", f1_score(testy, SVpred, average='weighted'))


yLR.append(scoreLR)
yNB.append(scoreNB)
yKNN.append(scoreKNN)
yDT.append(scoreDT)
ySVM.append(scoreSV)
```
